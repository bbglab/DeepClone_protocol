## Extended Data Figure 2: Comparison of deepUMIcaller and DupCaller across 8 normal bladder urothelium samples

This notebook is used to generate the following plots:

**g** Mutational profile of indels identified by DupCaller (top) or deepUMIcaller (bottom).

**h** Extra plot showing the mutation density for indels in deepUMIcaller and DupCaller


```python
# --- 1. Setup: imports -------------------------------------------
import sys, os
import pandas as pd
import matplotlib.pyplot as plt
import sigProfilerPlotting as sigPlt
```

# SigProfilerPlotting

```python
indels_matrix = pd.read_table("../data/samples.ID28.all")
indels_matrix = indels_matrix.set_index('MutationType')
indels_matrix
```

```python
indels_matrix.columns
```

```python
dupcaller_samples = [x for x in indels_matrix.columns if 'dupcaller' in x]
deepUMIcaller_samples = [x for x in indels_matrix.columns if x not in dupcaller_samples]
```

```python
indels_matrix["DupCaller"] = indels_matrix[dupcaller_samples].sum(axis=1)
indels_matrix["deepUMIcaller"] = indels_matrix[deepUMIcaller_samples].sum(axis=1)
```

```python
indels_matrix_groups = indels_matrix[["DupCaller", "deepUMIcaller"]].reset_index()
```

```python
indels_matrix_groups.to_csv("../data/samples.ID28.grouped", sep='\t', index=False)
```

### Run with sigprofiler_env

```python
sigPlt.plotID("../data/samples.ID28.grouped", "../plots/", "ID28_bladder", '28', percentage=False)
```
